```python
'''
문제 1. 
pandas 패키지와 seaborn 패키지를 import 하세요
'''

# 코드를 입력해주세요.
import pandas as pd
import seaborn as sns
```

```python
data1 = sns.load_dataset('attention')
data1
```

```python
'''
문제 2.
attention 내장 데이터셋을 불러왔습니다. 데이터셋에서 필요 없는 칼럼은 'Unnamed: 0'입니다.
인덱싱을 이용하여 Unnamed:0 칼럼을 제거해주세요. (칼럼명 입력 X)
칼럼이 제거된 데이터프레임을 data1에 다시 반영해주세요.
'''

# 코드를 입력해주세요.
data1 = data1.drop('Unnamed: 0', axis =1 )
data1
```

```python
'''
문제 3.
행 기준, 뒤에서 30번째 행부터 끝까지, 열 기준 1 인덱스 이후부터 끝까지 해당하는 
데이터 프레임을 data1에 반영해주세요.
'''

# 코드를 입력해주세요.
data1.iloc[-30:, 1:]
```

```python
from numpy.lib.arraysetops import unique
'''
문제 4.
for loop를 이용하여 각 칼럼에 존재하는 unique값을 출력해주세요.
- 형식 : 칼럼명 : [unique 값들]
정답 예시 : solutions : [1 2 3]
'''

# 코드를 입력해주세요.

for column in data1.columns:
  print(f'{column} : {data1[column].unique()}')


```

```python
'''
문제 5.
score(점수)가 6점 이상인 데이터의 행 개수를 인덱싱을 이용하여 출력하세요 (print)
'''
#인덱싱 이용이니까
# 코드를 입력해주세요.
print(data1.loc[data1['score'] >= 6].shape[0])
```

```python
'''
문제 6.
score(점수)가 6점 미만이고 attention 유형이 divided인 데이터를 추출해주세요.
'''

# 코드를 입력해주세요.
data1.loc[(data1['score'] < 6) & (data1['attention'] == 'divided')]
```

```python
'''
문제 7.
5번 문제의 데이터에서 attention 칼럼을 새로운 index로 세팅하세요.
'''
# 코드를 입력해주세요.

data1.loc[(data1['score'] < 6) & (data1['attention'] == 'divided')].set_index('attention')
```
