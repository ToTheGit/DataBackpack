#### boxplot의 의미
- box가 나타내는 것 : 1사분위수와 3사분위수 사이의 분포
- box 내부 가로선 : 중앙값(median)
- lower whisker : Q1 - 1.5 * IQR 보다 큰 값 중 최소값
- upper whisker : Q3 + 1.5 * IQR 보다 작은 값 중 최대값
- IQR : Q3 - Q1
- dot : outlier (lower whisker와 upper whisker사이 포함되지 않는 값)
