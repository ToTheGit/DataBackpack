```python
import gc
import glob
import geopandas as gpd

# SHAPE 파일 경로 리스트
file_path = glob.glob("C:\\Users\\nws71\\Downloads\\서울시 주요 50장소 영역 (1).zip")[0]

# Shape 파일을 GeoDataFrame으로 불러오기
seoul = gpd.GeoDataFrame.from_file(file_path, encoding='UTF-8')

# 경위도 좌표계 설정
seoul = seoul.to_crs({"init": "epsg:4326"})

# 데이터 정보
seoul.info()

# 데이터 예시
seoul.tail()

```
